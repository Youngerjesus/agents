```python
def get_html_body(website_url: str) -> str:
    cookies: Optional[dict] = None
    headers: Optional[dict] = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9",
        "Accept-Language": "en-US,en;q=0.9",
        "Referer": "https://www.google.com/",
        "Connection": "keep-alive",
        "Upgrade-Insecure-Requests": "1",
    }

    page = requests.get(
        website_url,
        timeout=15,
        headers=headers,
        cookies=cookies if cookies else {},
    )

    page.encoding = page.apparent_encoding
    parsed = BeautifulSoup(page.text, "html.parser")
    
    # 불필요한 태그 제거
    for tag in parsed.find_all(['script', 'style', 'nav', 'footer']):
        tag.decompose()

    # body 태그 찾기
    body = parsed.find('body')

    return body
```

```python
soup = BeautifulSoup(page.text, "html.parser")

# 불필요한 태그 제거
for tag in soup.find_all(['script', 'style', 'nav', 'footer']):
    tag.decompose()

# body 태그 찾기
body = soup.find('body')

print(body)
```

```python
def extract_title(soup: BeautifulSoup) -> str:
    try:
        # h1 태그에서 특정 클래스를 찾음
        h1_tag = soup.find('h1', class_='np_18px')
        if h1_tag:
            # h1 태그 안의 특정 클래스를 가진 span 태그를 찾음
            span_tag = h1_tag.find('span', class_='np_18px_span')
            if span_tag:
                return span_tag.get_text(strip=True)
            
    except Exception as e:
        print(f"Error extracting title: {e}")
        return "Error extracting title"

title = extract_title(soup)
print(title)
        
```

```python
def extract_content(soup: BeautifulSoup) -> dict:
    try:
        # article 태그 내의 xe_content 클래스를 가진 div 찾기
        article = soup.find('article')
        if not article:
            return {"content": "", "images": [], "error": "Article tag not found"}
            
        content_div = article.find('div', class_='xe_content')
        if not content_div:
            return {"content": "", "images": [], "error": "Content div not found"}
        
        # 텍스트와 이미지를 저장할 리스트
        text_parts = []
        image_urls = []
        
        # 모든 요소를 재귀적으로 순회 (중첩된 태그까지 처리)
        for element in content_div.descendants:
            if isinstance(element, str) and element.strip():
                # 순수 텍스트 노드 처리
                text = element.strip()
                if text:
                    text_parts.append(text)
                    
            elif element.name == 'img':
                # 이미지 URL 추출 (div 안에 있는 이미지도 포함)
                img_url = element.get('src', '')
                if img_url:
                    # 상대 경로를 절대 경로로 변환 (필요한 경우)
                    if img_url.startswith('/'):
                        img_url = f"https://{img_url}"
                    image_urls.append(img_url)
                    text_parts.append(f"[Image: {img_url}]")
            
            elif element.name == 'br':
                # 줄바꿈 처리
                text_parts.append('\n')
        
        # 텍스트 정리
        content = ' '.join(text_parts)
        content = re.sub('[ \t]+', ' ', content)
        content = re.sub('\\s+\n\\s+', '\n', content)
        content = re.sub('\n\s*\n', '\n\n', content)  # 연속된 빈 줄 처리
        
        return {
            "content": content.strip(),
            "images": image_urls,
            "error": None
        }
        
    except Exception as e:
        return {
            "content": "",
            "images": [],
            "error": f"Error extracting content: {str(e)}"
        }

# 디버깅을 위한 확장 버전
def extract_content_debug(soup: BeautifulSoup, debug: bool = True) -> dict:
    try:
        article = soup.find('article')
        if debug and not article:
            print("Article tag not found")
            return {"content": "", "images": [], "error": "Article tag not found"}
            
        content_div = article.find('div', class_='xe_content')
        if debug and not content_div:
            print("Content div not found")
            print("Available divs in article:", 
                  [div.get('class', []) for div in article.find_all('div')])
            return {"content": "", "images": [], "error": "Content div not found"}
            
        if debug:
            print(f"Found {len(list(content_div.descendants))} descendant elements")
            print(f"Found {len(content_div.find_all('img'))} images")
            
        # ... (이하 동일)
        
    except Exception as e:
        if debug:
            print(f"Error: {str(e)}")
            print("HTML structure:", soup.prettify())
        return {
            "content": "",
            "images": [],
            "error": f"Error extracting content: {str(e)}"
        }
    
content = extract_content(soup)
print(content)
```
